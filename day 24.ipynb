```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
df = pd.read_csv('Dataset/Titanic-Dataset.csv')

df.head()
```

one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

#Display initial categorical columns
cat_cols = df.select_dtypes(include=['object'])

#Make a copy of the original 'Sex' column
df_copy=cat_cols[['Sex']].copy()

#initialize and fit OneHotEncoder
enc = OneHotEncoder(sparse_output=False)
encoded_array = enc.fit_transform(df_copy)

#Get the encoded category names
encoded_feature_names = enc.get_feature_names_out(['Sex'])

# convert encoded array to dataframe with column names
encoded_df = pd.DataFrame(encoded_array, columns=encoded_feature_names)
# concatnate original and encoded columns
df_final = pd.concat([df_copy, encoded_df], axis=1)
## Show the result
df_final.head()
```

```python
studentdf=pd.read_csv("Dataset/university_students_list.csv")
studentdf.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Select the 'Course' column for ordinal encoding
course_col = studentdf[['Course']]

# Initialize and fit OrdinalEncoder
ordinal_enc = OrdinalEncoder()
studentdf['course_encoded'] = ordinal_enc.fit_transform(course_col)

# Show the result
studentdf[['Course', 'course_encoded']]
```
