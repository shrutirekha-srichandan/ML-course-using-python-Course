## Principal component analysis (PCA).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data=load_breast_cancer()
X=data.data
y=data.target
target_names=data.target_names
feature_names=data.feature_names
```

```python
X
```

```python
scalar=StandardScaler()
x_scaled=scalar.fit_transform(X)
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
# Step 4 : Visualize the first two principal components

plt.Figure(figsize=(10,7))
for target,color,label in zip([0,1],['red','blue'],target_names):
    plt.scatter(x_pca[y==target,0],x_pca[y==target,1],alpha=0.7,color=color,label=label)
    
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Breast Cancer Dataset")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
