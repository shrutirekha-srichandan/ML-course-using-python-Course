```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fmri = sns.load_dataset("fmri")
fmri.head(10)
```

```python
fmri
```

```python
sns.lineplot(data = fmri,x = "timepoint", y = "signal",style = "event",hue ="region")
plt.show()

#hue is used to color the lines based on "region" column
#style is used to differentiate the lines based on "event" column
```

```python
sns.lineplot(data = fmri,x = "timepoint", y = "signal",style = "event",hue ="region",markers=True, dashes=False)
plt.show()
#markers=True adds markers to the lines
#dashes=False removes the dashed lines, making them solid
```

```python
sns.lineplot(data = fmri,x = "timepoint", y = "signal",style = "event",hue ="region",markers=True, dashes=True)
plt.show()
#markers=True adds markers to the lines
#dashes=True allows the lines to be dashed, which can help differentiate between multiple lines in the plot
```

```python
#generate a sample pokemon dataset and save it as an Excel file
pokemon_data = {
    "Name": ["Pikachu", "Charmander", "Squirtle", "Bulbasaur", "Eevee"],
    "Type": ["Electric", "Fire", "Water", "Grass", "Normal"],
    "HP": [35, 39, 44, 45,60],
    "Attack": [55, 52, 48, 49,40],
    "Defense": [40, 43, 65, 49,45],
    "Speed": [45,65,43,90,20]
}
pokemon_df = pd.DataFrame(pokemon_data)
pokemon_df.to_excel("Dataset/pokemon_data.xlsx", index=False)
# Load the PokÃ©mon dataset from the Excel file
pokemon_df.head()
```

```python
sns.lineplot(data = pokemon_df, x="Attack", y ="Defense",hue = "Type", style = "Type", markers=True, dashes=False)
plt.show()
```

```python
sns.scatterplot(data = pokemon_df, x="Attack", y ="Defense",hue = "Type", style = "Type")
plt.show()
```

```python
sns.barplot(data = pokemon_df, x="Type", y ="Defense")
plt.show()
```

```python
sns.set_style("whitegrid")
colors = sns.color_palette("Set1", n_colors=5)
sns.barplot(data = pokemon_df, x="Name", y ="HP", palette = colors ,hue = "Name")
plt.legend(title="Pokemon Name", loc='upper left')  # Add a legend for clarity
#palette="viridis" sets the color palette for the bars
#palette="pastel" sets a pastel color palette for the bars
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.title("Pokemon HP comparison by Type")
plt.xlabel("Pokemon Name")
plt.ylabel("HP")
plt.show()
```

```python
# Load the Iris dataset
iris = sns.load_dataset("iris")
# Display the first few rows of the dataset
iris
```

```python
sns.scatterplot(data = iris, x="sepal_length", y="sepal_width", hue="species", style="species")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")  
plt.ylabel("Sepal Width")
plt.legend(title="Species")
plt.show()
```

```python
# generate a  sample diamond dataset 

diamond_data = {
    "Carat": [0.23, 0.21, 0.23, 0.29, 0.31, 0.24, 0.24, 0.26, 0.22, 0.23],
    "Cut": ["Ideal", "Premium", "Good", "Very Good", "Fair", "Ideal", "Premium", "Good", "Very Good", "Fair"],
    "Color": ["E", "E", "I", "J", "H", "E", "I", "J", "H", "G"],
    "Clarity": ["SI1", "SI2", "VS1", "VS2", "I1", "SI1", "SI2", "VS1", "VS2", "I1"],
    "Price": [326, 326, 327, 334, 335, 336, 337, 338, 339, 340],
}
diamond_df = pd.DataFrame(diamond_data)
diamond_df.head()
```

```python
sns.histplot(diamond_df['Carat'],kde =True, bins=10)
plt.title("Distribution of Diamond Carat")
plt.xlabel("Carat")
plt.ylabel("Density")
plt.show()
```

```python
sns.jointplot(data= iris,x ="sepal_length",y="sepal_width",hue="species"
              ,kind ="scatter"
              ,markers=["O","S","D"])
# This will create a joint plot with scatter points for each species
#markers=["O","S","D"] specifies different markers for each species
plt.suptitle("Joint Plot of Sepal Length and Sepal Width by Species", y=1.02)
#suptitle is used to set the title of the joint plot
plt.show()
```

```python
# Generate a sample churn dataset
churn_data = {
    "CustomerID": [f"CUST{i:03d}" for i in range(1, 21)],
    "Gender": np.random.choice(["Male", "Female"], 20),
    "Age": np.random.randint(18, 65, 20),
    "Tenure": np.random.randint(1, 10, 20),
    "Balance": np.round(np.random.uniform(1000, 10000, 20), 2),
    "NumOfProducts": np.random.randint(1, 4, 20),
    "HasCrCard": np.random.choice([0, 1], 20),
    "IsActiveMember": np.random.choice([0, 1], 20),
    "EstimatedSalary": np.round(np.random.uniform(20000, 100000, 20), 2),
    "Churn": np.random.choice([0, 1], 20)
}
churn_df = pd.DataFrame(churn_data)
churn_df.head()
```

```python
sns.violinplot(data=churn_df,x="Gender",y="Age",hue="Churn",split=True)
plt.title("Age Distribution by Gender and Churn Status")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
churn = pd.read_csv("Dataset/Churn Dataset.csv")
churn.head()
```

```python
sns.violinplot(data=churn,x="gender",y="age",hue="churn",split=True)
plt.title("Age Distribution by Gender and Churn Status")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#Boxplot on the first subplot
sns.boxplot(data=churn,x="gender",y="age",ax=axes[0]) #churn is the dataset
axes[0].set_title("Box plot of Age by Gender")
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Age")

#violinplot on the second subplot
sns.violinplot(data=churn,x="gender",y="age",hue="churn",split=True ,ax=axes[1])
axes[1].set_title("Violin plot of Age by Gender")
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("Age")

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=churn_df,x="Gender",y="Age") #churn_df is the dataset
plt.title("Age Distribution by Gender")
plt.xlabel("Gender")   
plt.ylabel("Age")
plt.show()
```
