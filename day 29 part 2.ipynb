### Decision Tree of Weather Data using Regressor

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
data = {
    'Temperature':[30,25,28,21,24,20,27],
    'Humidity':[65,70,68,75,80,78,83],
    'Windy':[0,0,1,1,0,1,1],
    'Play':[0,0,1,1,1,0,1]
}
df = pd.DataFrame(data)
print(df)
```

```python
#split features and targets
x = df.drop('Play',axis=1)
y = df['Play']
print('x:\n',x) #x
print('\n\ny:\n',y) #y
```

```python
# Train Desicion tree regressor
regressor = DecisionTreeRegressor(criterion='squared_error',random_state=42)
regressor.fit(x,y)
```

```python
#make a prediction
new_data =pd.DataFrame([[58,72,6]],columns=['Temperature','Humidity','Windy'])
prediction = regressor.predict(new_data)
print(prediction)
```

```python
# visualize the tree 
plt.figure(figsize=(7,7))
tree.plot_tree(regressor,feature_names=x.columns.tolist(),class_names=['no','yes'],filled=True,rounded=True)
plt.show()
```
