```python
import numpy as np
import numpy.random as rand
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10)
y=2*x
plt.plot(x,y)
```

## Add title and Labels

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## Changing line Aesthetics

```python
plt.plot(x,y,color='red',linestyle='dotted',linewidth=2)
plt.show()
```

```python
# Multiple Line graph
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,label='y=2x',color='blue',linewidth=2)
plt.plot(x,y2,label='y=3x',color='green',linewidth=3)
plt.title('Multiple Lines')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x=np.arange(0,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.title('Graph Subplot 1')
plt.plot(x,y1,label='y=2x',color='blue',linewidth=2)
plt.subplot(1,2,2)
plt.title('Graph Subplot 2')
plt.plot(x,y2,label='y=3x',color='green',linewidth=3)
plt.show()
```

## Bar Plot

```python
students={'John':25,'Alice':20,'Bob':18,'Diana':32}
names=list(students.keys())
ages=list(students.values())
plt.title('Bar Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.bar(names,ages,color='purple',width=0.3)
# plt.bar(names,ages,color=(0.5,0.9,0.7),width=0.3) # RGB color values
plt.show()
```

## Horizontal Bar Plot

```python
plt.barh(names,ages,color='orange')
plt.title('Horizontal Bar Chart')
plt.xlabel('Ages')
plt.ylabel('Names')
# plt.grid(True)
plt.show()
```

```python
x=np.arange(0,20)
y=rand.randint(1,100,size=20)
plt.title('Scatter Plot')
plt.scatter(x,y,color='red',marker='4',s=100)
```

```python
x=np.arange(0,10)
a1=rand.randint(1,100,size=10)
b1=rand.randint(1,100,size=10)
plt.title('Multiple Scatter Plots')
plt.scatter(x,a1,color='blue',label='Dataset A',marker='o',s=50)
plt.scatter(x,b1,color='red',label='Dataset B',marker='*',s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## Scatter plot subplot

```python
x=np.arange(0,10)
a1=rand.randint(1,100,size=10)
b1=rand.randint(1,100,size=10)
plt.subplot(1,2,1)
plt.scatter(x,a1,color='blue',label='Dataset A', marker='o',s=50)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.subplot(1,2,2)
plt.scatter(x,b1,color='red',label='Dataset B', marker='*',s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## Histogram

```python
data=rand.randn(1000)
plt.hist(data,bins=5,color='brown',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Box Plot

```python
one=rand.randint(1,100,size=10)
two=rand.randint(1,100,size=10)
three=rand.randint(1,100,size=10)
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

## Violin Plot

```python
plt.title('Violin Plot')
plt.violinplot([one,two,three],showmeans=True)
plt.show()
```

## Pie Plot

```python
fruits=['Apple','Banana','Cherry','Date']
quantities=[10,15,7,12]
plt.pie(quantities,labels=fruits)
plt.title('Pie Chart')
plt.show()
```

## Doughnut Chart

```python
plt.pie(quantities,labels=fruits,radius=3)
plt.title('Pie chart with Custom Radius')
plt.pie([1],colors='white',radius=1)
plt.show()
```
